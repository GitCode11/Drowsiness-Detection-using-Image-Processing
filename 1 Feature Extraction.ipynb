```python
import numpy as np
import math
import os.path
import imageio
from mlxtend.image import extract_face_landmarks
```

```python
def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear
```

```python
def mouth_aspect_ratio(mouth):
    A = distance.euclidean(mouth[14], mouth[18])
    C = distance.euclidean(mouth[12], mouth[16])
    mar = (A ) / (C)
    return mar
```

```python
def circularity(eye):
    A = distance.euclidean(eye[1], eye[4])
    radius  = A/2.0
    
    Area = math.pi * (radius ** 2)
    p = 0
    p += distance.euclidean(eye[0], eye[1])
    p += distance.euclidean(eye[1], eye[2])
    p += distance.euclidean(eye[2], eye[3])
    p += distance.euclidean(eye[3], eye[4])
    p += distance.euclidean(eye[4], eye[5])
    p += distance.euclidean(eye[5], eye[0])
    return 4 * math.pi * Area /(p**2)
```

```python
def mouth_over_eye(eye):
    ear = eye_aspect_ratio(eye)
    mar = mouth_aspect_ratio(eye)
    mouth_eye = mar/ear
    return mouth_eye
```

```python
def getFrame(sec,start=180000):
    vidcap.set(cv2.CAP_PROP_POS_MSEC, start + sec*1000)
    hasFrames,image = vidcap.read()
    return hasFrames, image
```

```python
# main program
from scipy.spatial import distance
import cv2
import os.path

k=120

for j in range(1,61):
    for i in range(0,11,10):
        if i == 10:
            checkpath = 'C:/Users/Sagar/Documents/Cudancudnn/Respondent_' + str(j) + '_0.csv'
            if not os.path.isfile(checkpath):
                print('Skipped ' + str(j) + '_' + str(i))
                k = k-1
                continue
            
        vidcap = cv2.VideoCapture('D:/Minor/Dataset/' + str(j) +'/' + str(i) + '.mp4')
        print('Traversing D:/Minor/Dataset/' + str(j) +'/' + str(i) + '.mp4')
        
        data = []
        labels = []
        
        sec = 0
        frameRate = 1
        success, image =getFrame(sec)
        count = 0
        flag=0
        
        while success and count < 240:
            
            landmarks = extract_face_landmarks(image)
            
            if landmarks is None:
                
                if flag == 1:
                    print('Not found in '+str(j)+'_'+str(i))
                    flag = 2
                    k = k-1
                    break
                    
                sec = sec + frameRate
                sec = round(sec, 2)
                success, image = getFrame(sec,start=180000)
                flag = 1
                
            elif sum(sum(landmarks))!=0:
                count += 1
                data.append(landmarks)
                labels.append([i])
                sec = sec + frameRate
                sec = round(sec, 2)
                success, image = getFrame(sec)

            else:
                sec = sec + frameRate
                sec = round(sec, 2)
                success, image = getFrame(sec)
                print("Not Detected")
                
        if flag == 2:
            continue
            
        data = np.array(data)
        labels = np.array(labels)

        features = []

        for d in data:
            eye = d[36:68]
            ear = eye_aspect_ratio(eye)
            mar = mouth_aspect_ratio(eye)
            cir = circularity(eye)
            mouth_eye = mouth_over_eye(eye)
            features.append([int(j),ear, mar, cir, mouth_eye,int(i)])

        features = np.array(features)

        np.savetxt('Respondent_'+str(j)+'_'+str(i)+'.csv', features, delimiter = ",")
```
